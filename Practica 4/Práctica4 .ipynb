```python
def consulta1(cur):
    print ("Años que tengan una película que recibió una calificación de 4 o 5")
    cur.execute("SELECT DISTINCT Año FROM Clasificación NATURAL JOIN Películas WHERE (Estrellas = 4 OR Estrellas = 5) ORDER BY Año" )
    for fila in cur.fetchall():
        print (fila[0])
```

```python
def consulta2(cur):
    print ("Nombres de todos los críticos que hayan contribuido con tres o más calificaciones")
    cur.execute("SELECT DISTINCT Nombre FROM Críticos NATURAL JOIN Clasificación GROUP BY Idc HAVING COUNT(*) > 2" )
    for fila in cur.fetchall():
        print (fila[0])
```

```python
def consulta3(cur):
    print ("Títulos de películas y clasificaciones promedio")
    cur.execute("SELECT DISTINCT Título, AVG(Estrellas) FROM Películas NATURAL JOIN Clasificación GROUP BY Título ORDER BY AVG(Estrellas) DESC, Título" )
    for fila in cur.fetchall():
        print(fila[0], " =>")
        est = "%.2f" % fila[1]
        print(est, " *****")
```

```python
def consulta4(cur):
    print ("Mayor cantidad de estrellas que recibió la película")
    cur.execute("SELECT DISTINCT Título, MAX(Estrellas) FROM Películas NATURAL JOIN Clasificación GROUP BY Título ORDER BY Título" )
    for fila in cur.fetchall():
        print(fila[0], "=>")
        print("max : ", fila[1])
```

```python
def consulta5(cur):
    print ("Películas (media >= 4 estrellas), 25 años mas al lanzamiento")
    cur.execute("SELECT DISTINCT Título, Año + 25 FROM Películas NATURAL JOIN Clasificación GROUP BY Título HAVING AVG(Estrellas) >= 4" )
    for fila in cur.fetchall():
        print(fila[0], "=>")
        print("Año : ", fila[1])
```
